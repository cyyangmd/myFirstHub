```python



import openpyxl

# 打开 Excel 文件
workbook = openpyxl.load_workbook('/Users/user/Desktop/業務及權限角色列表.xlsx')

# 选择指定工作表
sheet = workbook['權限管理系統角色列表']

# 遍历每一行
for row_number, row in enumerate(sheet.iter_rows(values_only=True), start=1):
    # 遍历每一列
    for column_number, value in enumerate(row, start=1):
        # 打印行列位置和值
        print(f"Row {row_number}, Column {column_number}: {value}")
```

```python
import openpyxl

# 打开 Excel 文件
workbook = openpyxl.load_workbook('/Users/user/Desktop/業務及權限角色列表.xlsx')

# 选择指定工作表
sheet = workbook['權限管理系統角色列表']

# 遍历每一行
for row in sheet.iter_rows(values_only=True):
    # 打印每一行的值及其位置
    for i, value in enumerate(row, start=1):
        print(f"Column {i}: {value}")
```
